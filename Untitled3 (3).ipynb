```python
import pandas as pd

# (a) What is the overall number fatal accidents for the entire dataset period from 1975-2020? (Note: the denomenator is the total number of accidents)



number_of_fatal_accidents = 0 
x = range(1975, 2020, 5)

for year in x:
    
   
   df = pd.read_csv(fr"C:\Users\abhis\Downloads\Asst\accidentallyears\{year}ACCIDENT.CSV" ) 
   number_of_fatal_accidents += df["FATALS"].sum()
    
print("Total number of fatal accidents: ", number_of_fatal_accidents)    

```

```python
# b) How many people died over that period? How many total people were involved (fatal and non-fatal)?


people_died = 0
people_involved = 0
x =  range(1975, 2020, 5)

for year in x:
    df = pd.read_csv(fr'C:\\Users\\abhis\\Downloads\\Asst\\accidentallyears\\{year}ACCIDENT.CSV')
    people_died += df["FATALS"].sum()
    people_involved += df["PERSONS"].sum()

    
print("number of people died: ", people_died)
print("number of people involved: ", people_involved)   
  
```

```python
# What proportion of accidents occurred between 9pm and 4am (overnight)?

accidents = 0
accidents_overnight = 0
for year in x:
    df = pd.read_csv(fr'C:\\Users\\abhis\\Downloads\\Asst\\accidentallyears\\{year}ACCIDENT.CSV')
    accidents += df.shape[0]
    accidents_overnight += df[(df["HOUR"] >= 21) | (df["HOUR"] <= 4)].shape[0]

print("Total number of accidents: ", accidents)
print("Total number of accidents occurred between 9pm and 4am: ", accidents_overnight)
print("Proportion of accidents occurred between 9pm and 4am: ", accidents_overnight/accidents)
```

```python
# What proportion of accidents occurred when the weather was snowy?

accidents = 0
accidents_snowy = 0
for year in x:
    df = pd.read_csv(fr'C:\\Users\\abhis\\Downloads\\Asst\\accidentallyears\\{year}ACCIDENT.CSV')
    accidents += df.shape[0]
    accidents_snowy += df[df["WEATHER"] == 4].shape[0]

print("Total number of accidents: ", accidents)
print("Total number of accidents occurred when the weather was snowy: ", accidents_snowy)
print("Proportion of accidents occurred when the weather was snowy: ", accidents_snowy/accidents)
```

```python
# Build a line graph that shows the total fatalities by year (using the dataset with just every 5 years of data).

import matplotlib.pyplot as plt

total_fatalities = []
x = range(1975,2020,5)

for year in x:
    df = pd.read_csv(fr'C:\\Users\\abhis\\Downloads\\Asst\\accidentallyears\\{year}ACCIDENT.CSV')
    total_fatalities.append(df["FATALS"].sum())

plt.plot(x, total_fatalities)
plt.xlabel("Yrs")
plt.ylabel("fatal deaths")
plt.title("fatal deaths by year")
plt.show()
```

```python
final_data = df.sample(5000)
```

```python

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np


df = final_data.copy()

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)

dist = euclidean_distances(df_scaled, df_scaled)

distance_table = pd.DataFrame(dist, columns=final_data.index, index=final_data.index)
distance_table.head()
```

```python
df["nearest_neighbors"] = df.index.map(lambda x: distance_table[x].index[1:21].tolist())
df
```

```python

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


pca = PCA()

pca.fit(df_scaled)

explained_variance = pca.explained_variance_ratio_

cummulative_variance = np.cumsum(explained_variance)

plt.plot(cummulative_variance)
plt.xlabel("number of components")
plt.ylabel("cummulative explained variance")
plt.show()

print(f"The number of components necessary to capture 90% of the data is {np.argmax(cummulative_variance > 0.9) + 1}.")
```

```python
#  first component
first_component = pca.components_[0]
# Index of 1st component
largest_value_index = np.argmax(first_component)
# get the feature name
feature_name = df.columns[largest_value_index]
print(f"The feature that dominates the first component is {feature_name}.")

```
